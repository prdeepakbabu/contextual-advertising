```python
import logging
import os.path
import sys
import multiprocessing
import cython
import gensim

from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
```

```python
#import 5M NSA titles (also preprocessed for variations in case, lemmatization)
import cPickle as pickle
with open("C:\\Users\\badeepak\Documents\Work\URL PC\\pickled_objects\cleaned_title", 'rb') as fp:
  [cleaned_title] = pickle.load(fp)
fp.close()
```

```python
tmp = [i.split(' ') for i in cleaned_title]
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
sentences = [['first', 'sentence'], ['second', 'sentence']]
model = Word2Vec((tmp), min_count=3,size=300,workers=multiprocessing.cpu_count(),window=5)
model.init_sims(replace=True)
model.wv.save_word2vec_format("nsa_title_embedding_300",binary=True)
```

```python
print model.most_similar("walmart")
print model.most_similar("india")
print model.similarity("green","blue")
print model.similarity("man", "woman")
print model.most_similar("chromecast")
print model.wv.most_similar(positive=['woman', 'king'], negative=['man'])
print model.wv.most_similar(positive=['steroid', 'iphone'], negative=['drug'])
print model.wv.most_similar(positive=['smartwatch', 'drone'], negative=['pebble'])
```
